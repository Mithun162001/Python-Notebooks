# PANDAS BUILT IN DATA VISUALIZATION

```python
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df1= pd.read_csv('D:\Datasets\df1', index_col= 'Unnamed: 0')
df2= pd.read_csv('D:\Datasets\df2')
```

```python
df1.head()
```

```python
df2.head()
```

*1. Histogram*

```python
df1['A'].plot.hist()    # this looks like matplotlib histogram since seaborn is not imported
```

```python
df1['A'].hist()     # it's not recommended to use plot if we you want to take advantage of seaborn
```

```python
# ANother way of plotting
df1['A'].plot(kind= 'hist')
```

*2. Areaplot*

```python
df2.plot.area(figsize=(8,5), cmap= 'seismic')
```

*3. Barplot*

```python
df2.plot.bar(figsize= (6,5))    # Simple bar plot or simple bar diagram
```

```python
df2.plot.bar(stacked= True, figsize= (5,5))     # Multiple bar diagram
```

*4. Lineplot*

```python
df1.plot.line(x='A', y='B', cmap= 'Accent')
```

*5. Scatterplot*

```python
df1.plot.scatter(x='A', y='B')
```

```python
df1.head()
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100, alpha=0.5)   # This is similar to bubble plot
# here value of C is based upon the size
```

*6. boxplot*

```python
df2.boxplot()
```

*7. kdeplot*

```python
df2.plot.kde()
```

# NOTE:-
* Pandas built in data visualizations are not recommended to use. 
* They are not customizable
* They should be used only for references purpose and not as main plots to display
* Generally, we only use this for histograms and lineplot only.
