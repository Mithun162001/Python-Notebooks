```python
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp= [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

# Functional approach

** Normal Line Chart **

```python
plt.plot(years, gdp, color= 'black', marker= '.', linestyle= 'solid')
plt.xlabel("Years")
plt.ylabel('GDP')
plt.title("Nominal GDP")
```

```python
x= np.arange(1,20,2)
print(x)
y= np.arange(0,20,2)
print(y)
plt.plot(x,y)
```

** Subplots **

```python
plt.subplot(2,2,1)              # plt.subplot(nrows, ncols, index)
plt.plot(x,y, color = 'Red')
plt.xlabel('Years')
plt.ylabel("GDP")
plt.title('Nominal GDP')

plt.subplot(2,2,2)
plt.plot(y,x, color= 'green')
plt.xlabel('GDP')
plt.ylabel("Years")
plt.title('Inverted Nominal GDP')

plt.subplot(2,2,3)
plt.plot(y,x, color= 'green')
plt.xlabel('GDP')
plt.ylabel("Years")
plt.title('Inverted Nominal GDP')

plt.tight_layout()
```

# Object Oriented Approach

** Simple Line chart using objects**

```python
# we will be using the figure object instead of usual matplotlib object
fig= plt.figure()
```

```python
type(fig)   # we can see that it has created a figure object of class Matplotlib
```

In simple words figure contains 0 axes which means it's an empty canvas.. 

To add axes into the figure/ empty canvas we use add_axes() method

```python
fig = plt.figure()  # empty figure is created with 0 axes

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('Year')  # set_xlabel() instead of xlabel because this is a different object and class which contains different methods
axes.set_ylabel('GDP')
axes.set_title('Nominal GDP')

# this means that we are using 10% left side of canvas, 10% bottom side of canvas, 80% width of canvas, 80% height of canvas
```

```python
fig= plt.figure()
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.22,0.5,0.4,0.3])

axes1.plot(years,gdp)
axes1.set_title("Outer wear")
axes1.set_xlabel("Years")
axes1.set_ylabel("GDP")

axes2.plot(gdp,years)
axes2.set_title("Inner wear")
axes2.set_ylabel("Years")
axes2.set_xlabel("GDP")
```

**Subplots**

```python
fig, axes = plt.subplots(nrows= 3, ncols=3)
plt.tight_layout()          # for having a better layout of graphs we use it
```

```python
# here it showed that axes is just like a array of lists containg the axes values of subplots
# Hence it can be traversed through the for loop
fig, axes= plt.subplots(nrows=1, ncols=2)
for i in axes:
    i.plot(years, gdp)
    plt.tight_layout()
```

```python
# Since axes is a array of lists it can even be indexed
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(years, gdp)
axes[0].set_xlabel('years')
axes[0].set_ylabel('GDP')
axes[0].set_title('Nominal GDP')

axes[1].plot(x,y)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Dummy Values')
plt.tight_layout()
```

# Figure Size/ Aspect ratio, DPI

```python
fig = plt.figure(figsize= (20,9), dpi= 32.4)    # DPI- Dots per inch 
ax= fig.add_axes([0,1,1,1])
ax.plot(years, gdp)
```

**For Subplots**

```python
fig, axes= plt.subplots(nrows=2, ncols=2, figsize= (3,3), dpi=100)
axes[0,0].plot(years,gdp)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)

plt.tight_layout()
```

# Legend

```python
fig= plt.figure()
axes= fig.add_axes([0,0,1,1])   # left, bottom, width, height 

axes.plot(x,x*2, color= 'red', label= 'x*2 plot')
axes.plot(x,x**2, color= 'purple', label = 'x**2 plot')
axes.legend(loc='best')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize= (8,5))
axes[0].plot(x,x**2, label= 'x square')
axes[0].plot(x,x**3, label = 'x**4', color= 'purple')
axes[1].plot(x, x**3, color= 'red', label= 'x cube')
axes[0].legend()
axes[1].legend()
```

**Plot Apperance**

We have many plot customizations available in matplotlib

like the color, linestyle, linewidth, alpha, marker, markersize, markerfacecolor and so on

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y)  # this is the default plot available, but this can be more customised
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize= (8,5))
axes.plot(x,y, color='black', linestyle='-', linewidth= 3, marker= 'o', markersize= 10, markerfacecolor= 'cyan', markeredgewidth=3, markeredgecolor= 'black')
axes.set_xlim([0,20])
axes.set_ylim([0,20])
axes.set_title('DATA SPACE TECH')
axes.set_xlabel('X axis')
axes.set_ylabel('y axis')
```
