# K MEANS CLUSTERING

```python
# Importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Here we shall use the artificial data which can be genrated in sklearn library
from sklearn.datasets import make_blobs
# make_blobs() is used to create artificial data for clustering
```

```python
data = make_blobs(n_samples=200, n_features= 2, centers= 4, cluster_std= 1.8, random_state= 101)
# n_samples ----> is the number of obseravtions needed in a data
# n_features ---> is the number of features/ columns in a data
# centers ------> is the number of centroid points in a data
# cluster_std --> is the standard deviation of the clsuter points
```

```python
data[0][:5,:]

# Here we see that data[0] has these 2 features which we had mentioned before
# and are having 200 sample data points as well
# the data created using make_blobs() has created tuples
# data[0] has the values which we had mentioned in make_blobs()
```

```python
data[1][:5]

# here the data[1] has the labelled data of values 0,1,2 since we had 3 centroids mentioned in params
# generally in real world data we will not have the labelled data like this
# for better understanding and to see the differences between clustered segmentation we are using here
```

*Clusters which we already knew of labelled artificial data*

```python
sns.scatterplot(x= data[0][:,0], y= data[0][:,1], hue = data[1], palette='Set1')
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')
```

```python
# we shall look at how our data looks like using plotting libraries
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap= 'rainbow_r')
```

*Clusters which we will see now from K Means Algorithm*

```python
# importing KMeans method from sklearn
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)

# Since we already know there are 3 clsuters in the data because we created 3 cenroids while creating the artificial data
```

```python
model.fit(data[0])

# here we fit the data of the datapoints created in the tuple
```

```python
# To see the centroid values 
model.cluster_centers_

# these are the datapoints of the centroids in 0,1,2 clusters respectively
```

```python
# To see the labels which is predicted from the KMeans
model.labels_

# These are the groups of points to a cluster
# 0 ---> 1st cluster

# 1 ---> 2nd cluster
# 1 ---> 3rd cluster

# The model has segmentated the cluster values for the datapoints in the data
```

```python
# now we shall see the scatterplot of the predicted one

sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue= model.labels_, palette='rainbow')
```

```python
# For better understanding how the segmented one and the labelled one looks like we can use the subplots
sns.set_style('white')
fig, (ax1,ax2) = plt.subplots(1,2,sharey= True, figsize= (10,6))

ax1.set_title('K Means Segmented')
ax1.scatter(data[0][:,0], data[0][:,1], c= model.labels_, cmap='Set1_r')

ax2.set_title('original clusters')
ax2.scatter(data[0][:,0], data[0][:,1], c= data[1], cmap='Set1_r')
```
