```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('D:\Datasets\CarPrice_Assignment.csv', index_col=0)
df[:5]
```

```python
df.columns
```

```python
df['enginelocation'].unique()
```

```python
df.drop(['symboling', 'CarName'], axis= 1, inplace= True)
```

```python
df.drop(['carbody', 'drivewheel'], axis= 1, inplace= True)
```

```python
df[:5]
```

```python
def labeling(fuel):
    if fuel == 'gas':
        fuel = 0
        return fuel
    else:
        fuel = 1
        return fuel
df['fueltype']=df['fueltype'].apply(lambda x: labeling(x))
```

```python
def labeling(fuel):
    if fuel == 'front':
        fuel = 0
        return fuel
    else:
        fuel = 1
        return fuel
df['enginelocation']=df['enginelocation'].apply(lambda x: labeling(x))
```

```python
def labeling(aspire):
    if aspire == 'std':
        aspire = 0
        return aspire
    else:
        aspire = 1
        return aspire
df['aspiration']=df['aspiration'].apply(lambda x: labeling(x))
```

```python
def labeling(door):
    if door == 'two':
        door = 0
        return door
    else:
        door = 1
        return door
df['doornumber']=df['doornumber'].apply(lambda x: labeling(x))
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=df.corr(), annot= True)
sns.set_style('whitegrid')
```

```python
X = df[['fueltype', 'aspiration', 'doornumber', 'enginelocation', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
X[:5]
```

```python
y= df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
pred_lm = lm.predict(X_test)
pred_lm[:5]
```

```python
y_test[:5]
```

```python
from sklearn import metrics
metrics.r2_score(y_test, pred_lm)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
pred_rf = rf.predict(X_test)
pred_rf[:5]
```

```python
y_test[:5]
```

```python
metrics.r2_score(y_test, pred_rf)
```
